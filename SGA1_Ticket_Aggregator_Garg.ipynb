1. <b>Data Organisation:</b> All the four variables will further pass on the data to a <code>Python List - **data**</code>.
    1. <code>data[0]</code> will take up <code><i>ticket_price</i></code>

    1. Ticket Price: <code><i>tkt_price</i></code> function will be used to process and verify the variable <code><i>ticket_price</i></code>. It will ensure that only <code>float</code> value is entered. The entered value will be round up to 2 decimal place, using <code>round()</code>.
    1. Number of Connecting flights: <code><i>conn_flight</i></code> function will be used to process and verify the variable <code><i>c_flight</i></code>. It will ensure that only <code><i>int</i></code> value is entered.
    1. Refund availability: <code><i>refund</i></code> function will be used to process and verify the variable <code><i>refund_avail</i></code>. It will ensure that only **Yes/No** is entered.

1. <b>Data Categorization:</b> After all the data entry, verification and re-edit, if needed, we will use the <code>check_offer</code> function to ___Categorize___ the information in to the **four categories** provided through the assignment. We will use the <code>if-elif-else</code> function to achive the categorisation.

```python
def tkt_price(data):

    n = True

    # Instructions to facilitate user in making a correct ticket price.
    print(' 1. Enter the ticker price in Dollors (58 Dollor and 38 cents = 50.38)','\n',
          '2. 3 or more decimal numbers will be round off to 2 decimal value (20.102 = 20.10)')

    # ticket price will be stored as a first element in the list data 
    while n == True:
        try:
            ticket_price = float(input())
            data.append(round(ticket_price,2))
            n = False
        except ValueError as ve:
            print('Entered value is not a number. Please enter the correct ticket price.')
    
def conn_flight(data):
    n=True
    # Instructions to facilitate user in making a correct number of connecting flight.

    print(' 1. Enter the number of connecting flight as an integer','\n',
          '2. If a flight has 2 connecting flight, please enter "2"','\n',
         '3. Enter 0 for direct flight')

    # flight connection status will be stored as a second element in the list data 
    while n == True:
        try:
            c_flight = int(input())
            data.append(c_flight)
            n = False
        except ValueError as ve:
            print('Entered value is not an integer. Please enter the correct number of connecting flights.')

def refund(data):
    n=True
    # Instructions to facilitate user in making a correct refund facility.

    print('Is the refund included? (Yes/No)')

    # refund status will be stored as a second element in the list data 
    while n == True:
        refund_avail = input()
        if refund_avail == 'Yes' or refund_avail == 'No':
            data.append(refund_avail)
            n = False
        else:
            print('Please enter Yes or No')

def luggage(data):
    n=True
    # Instructions to facilitate user in making a correct luggage status.

    print('Is the luggage included? (Yes/No)')

    # luggage status will be stored as a second element in the list data 
    while n == True:
        luggage_allowed = input()
        if luggage_allowed == 'Yes' or luggage_allowed == 'No':
            data.append(luggage_allowed)
            n = False
        else:
            print('Please enter Yes or No')

def print_entries(data):
    if len(data)==4:
        print('You have entered the following information')
        print('1. Ticket price:\t\t${0:.2f}\n2. Number of Connecting Flights:{1:d}\n'
            '3. Refund Included:\t\t{2}\n3. Luggage Included:\t\t{3}'.format(data[0], data[1], data[2],data[3]))
        print('\n\n')
        answer = input('Are the entries correct (Yes/No)')
        if answer == 'Yes':
            print('\n')
        elif answer == 'No':
            print('Let us the entries again')
            data.clear()
            tkt_price(data)
            conn_flight(data)
            refund(data)
            luggage(data)
            print_entries(data)
        else:
            print_entries(data)
    else:
        print('First provide complete information regarding')
        print('1. Ticket price\n2. Number of Connecting Flights\n'
            '3. Refund Included\n3. Luggage Included')
def check_offer(data):
    best_r_l = ['Yes','Yes']

    if data[0] < 200 and data[1] <= 1 and data[2:] == best_r_l:
        print('The Best')
    elif data[0] >= 200 and data[0] <= 250 and data[1] <= 2:
        print('Good Enough')
    elif data[0] > 250 and data[1] >= 3:
        print('Bad')
    else:
        print('Other')
```
